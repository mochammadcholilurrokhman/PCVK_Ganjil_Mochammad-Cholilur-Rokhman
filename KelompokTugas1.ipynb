```python
import cv2
from skimage import io as skio
```

```python


# Load the image directly from the local path in Google Colab
image_path = "/KTP.jpg"  # Replace with the actual file path in Colab
image_ktp = skio.imread(image_path)

# Resize the image
image_ktp = cv2.resize(image_ktp, (0, 0), fx=0.5, fy=0.5)

# Convert the image from BGR to RGB
image_5 = cv2.cvtColor(image_ktp, cv2.COLOR_BGR2RGB)

# Concatenate the original and RGB images horizontally
final_frame = cv2.hconcat([image_ktp, image_5])

# Display the final image (uncomment if running in a Jupyter notebook)
from google.colab.patches import cv2_imshow
cv2_imshow(final_frame)

# Save the final image or display using OpenCV's imshow (for local use)
# cv2.imshow("Final Image", final_frame)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
image_5 = cv2.cvtColor(image_ktp, cv2.COLOR_BGR2RGB)
image_3 = cv2.cvtColor(image_ktp, cv2.COLOR_BGR2RGB)

tinggi_kotak_agama = 10
lebar_kotak_agama = 100

tinggi_kotak_status = 10
lebar_kotak_status = 100

start_y_agama = 140
start_x_agama = 95

start_y_status = 150
start_x_status = 95

for x in range(lebar_kotak_agama):
  for y in range(tinggi_kotak_agama):
    image_3[start_y_agama+y,start_x_agama + x] = [0,255,255]

for x in range(lebar_kotak_status):
  for y in range(tinggi_kotak_status):
    image_3[start_y_status+y,start_x_status + x] = [0,255,0]


final_frame = cv2.hconcat((image_5,image_3))
cv2_imshow(final_frame)
```
